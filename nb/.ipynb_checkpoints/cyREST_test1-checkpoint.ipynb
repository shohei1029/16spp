```python
import requests
import json
from py2cytoscape.data.cyrest_client import CyRestClient
import networkx as nx
from IPython.display import Image

cy = CyRestClient()
print(json.dumps(cy.status(), indent=4))
```

```python
# Basic Setup
PORT_NUMBER = 1234

# Specify your machine's URL (The IP address of the machine running Cytoscape and cyREST) if you use 
# Docker or remote server for this notebook.
#IP = '192.168.1.1'

# If you are running both Notebook server and Cytoscape on a same machine, just use localhost
IP = 'localhost'

BASE = 'http://' + IP + ':' + str(PORT_NUMBER) + '/v1/'

# Header for posting data to the server as JSON
HEADERS = {'Content-Type': 'application/json'}
```

```python
G=nx.Graph()
with open("/home/shohei/bio/16spp/analysis/sim/A,B,C_pi90_blastp_1e-5_HIV-1-gM-noRs_pol-aa_v3.sim.txt") as in_txt:
    for line in in_txt:
        line = line.rstrip()
        #source, edge_attr, target = line.split(" ")
        source, target, edge_attr = line.split(" ")

        G.add_edge(source, target, weight=float(edge_attr) )
```

```python
g_cy = cy.network.create_from_networkx(G)
print('network created: SUID = ' + str(g_cy.get_id()))
nid = g_cy.get_id()
```

```python
#create view
requests.post(BASE + "networks/{}/views/".format(g_cy.get_id()))
```

```python
layouts = cy.layout.get_all()
print(json.dumps(layouts, indent=4))

cy.layout.apply(name='force-directed', network=g_cy)
```

```python
#http://localhost:1234/v1/styles/
style_sn = cy.style.create('BioPAX_small_ds')
cy.style.apply(style_sn, network=g_cy)
```

```python
#get png 
network_png = requests.get(BASE + "networks/{}/views/first.png?h=2000".format(g_cy.get_id()))
#network_png = g_cy.get_png() #
```

```python
Image(network_png.content)
with open("/home/shohei/bio/16spp/results/networks/A,B,C_pi90_blastp_1e-5_HIV-1-gM-noRs_pol-aa_v3.png", "wb") as f:
    f.write(network_png.content)
```

```python
#destroy network
requests.delete(BASE + "networks/{}/".format(g_cy.get_id()))
```
