```r
library(ggplot2)
library(dtplyr) #data.table + dplyr code now lives in dtplyr.
#library(data.table)
#library(dplyr)
#library(reshape2)
```

```r
in_file <- "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/CRE_region_array.csv"
data <- read.table(in_file, header = T, sep = ",")
```

```r
str(data)
```

```r
region_cre.RVT_connect <- data.frame(x=1:length(data$CRE[data$region_name=="RVT_connect"]), y=data$CRE[data$region_name=="RVT_connect"])
region_cre.RNase_H <- data.frame(x=1:length(data$CRE[data$region_name=="RNase_H"]), y=data$CRE[data$region_name=="RNase_H"])
```

```r
print(region_cre.RVT_connect)
```

```r
#RVT_connect
p <- ggplot(region_cre.RVT_connect, aes(x=x, y=y))

###折れ線グラフ or 棒グラフ
#p <- p + geom_line()
p <- p + geom_bar(stat = "identity")
###

#p <- p + scale_x_continuous(limits = c(41,100))

p <- p + scale_x_continuous(limits = c(41, 100), breaks = c(41, 100), labels = c(353, 413), expand = c(0.01, 0.01))
p <- p + scale_y_continuous(breaks = c(0.00, 0.15, 0.25))

#p <- p + xlab("postions") + ylab("Cumulative relative entropy")
p <- p + xlab("") + ylab("")

p <- p + theme_classic(base_size = 16, base_family = "Helvetica") 
p <- p + theme(axis.title = element_text(face = "bold",size = rel(1)),
               axis.title.x = element_text(size = 20, family = "Helvetica"),
               axis.title.y = element_text(size = 20, family = "Helvetica"), 
               axis.text.x = element_text(size = 12, colour = 1, family = "Helvetica"), 
               axis.text.y = element_text(size = 12, colour = 1, family = "Helvetica"),
               aspect.ratio = 0.1,
              # plot.margin=unit(c(10,5,5,5),"mm"),
              )
#↑設定の重複とかあるけど勘弁。メモも兼ねてるから..。
plot(p)

# ggsave(plot=p, "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/fig_pos_CRE_region-RVT_connect_R_geom_line.png")
# ggsave(plot=p, "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/fig_pos_CRE_region-RVT_connect_R_geom_line.svg")
```

```r
#RNase_H
p <- ggplot(region_cre.RNase_H, aes(x=x, y=y))

###折れ線グラフ or 棒グラフ
p <- p + geom_line()
#p <- p + geom_bar(stat = "identity")
###

#p <- p + scale_x_continuous(limits = c(41,100))
p <- p + scale_x_continuous(limits = c(41, 100), breaks = c(41, 100), labels = c(476, 536), expand = c(0.01, 0.01))
p <- p + scale_y_continuous(breaks = c(0.00,0.15))

#p <- p + xlab("postions") + ylab("Cumulative relative entropy")
p <- p + xlab("") + ylab("")

p <- p + theme_classic(base_size = 16, base_family = "Helvetica") 
p <- p + theme(axis.title = element_text(face = "bold",size = rel(1)),
               axis.title.x = element_text(size = 20, family = "Helvetica"),
               axis.title.y = element_text(size = 20, family = "Helvetica"), 
               axis.text.x = element_text(size = 12, colour = 1, family = "Helvetica"), 
               axis.text.y = element_text(size = 12, colour = 1, family = "Helvetica"),
               aspect.ratio = 0.1,
              # plot.margin=unit(c(10,5,5,5),"mm"),
              )
#↑設定の重複とかあるけど勘弁。メモも兼ねてるから..。

plot(p)

# ggsave(plot=p, "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/fig_pos_CRE_region-RNase_H_R_geom_bar.png")
# ggsave(plot=p, "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/fig_pos_CRE_region-RNase_H_R_geom_bar.svg")
```

```r
#CRE Z-scores
in_file <- "../../analysis/CRE/sorted_removed-HXB2_mafft-linsi_with-HXB2_A,B,C_HIV-1-gM-noRs_pol-aa_v3/CRE-Z_region_array.csv"
data <- read.table(in_file, header = T, sep = ",")
region_cre.RVT_connect <- data.frame(x=1:length(data$CRE[data$region_name=="RVT_connect"]), y=data$CRE.Z[data$region_name=="RVT_connect"])
region_cre.RNase_H <- data.frame(x=1:length(data$CRE[data$region_name=="RNase_H"]), y=data$CRE.Z[data$region_name=="RNase_H"])
```
